```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CSVファイルを読み込む
df = pd.read_csv('history_pop.csv')
df
```

```python

# 世代ごとにグループ化
grouped = df.groupby('generation')

# 最大値、標準偏差、平均値を計算
x = list(grouped.groups.keys())
max_values = grouped['fitness'].max()
std_values = grouped['fitness'].std()
mean_values = grouped['fitness'].mean()


# グラフを作成
plt.figure(figsize=(10, 6))
plt.plot(max_values, label='Max', color='red')
plt.fill_between(x, std_values + mean_values, mean_values - std_values, alpha=0.2, color='blue', label='Std Dev')
plt.plot(mean_values, label='Mean', color='blue')
plt.xlabel('generation')
plt.ylabel('fitness')
plt.savefig("fitness_trends")
plt.show()
```
