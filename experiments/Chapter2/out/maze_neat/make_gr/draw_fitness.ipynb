```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CSVファイルを読み込む
df1 = pd.read_csv('history_fitness_ts1.csv')
df2 = pd.read_csv('history_fitness_semi.csv')
df3 = pd.read_csv('history_fitness_medium.csv')
```

```python
ts1 = df1["timestep"]
ts2 = df2["timestep"]
ts3 = df3["timestep"]



# グラフを作成
plt.figure(figsize=(10, 6))
plt.plot(ts1, label='split', color='red')
plt.plot(ts2, label='nomal', color='blue')
plt.plot(ts3, label='non-processing', color='yellow')
plt.xlabel('generation')
plt.ylabel('timestep')
plt.savefig("timestep_trends")
plt.show()
```
